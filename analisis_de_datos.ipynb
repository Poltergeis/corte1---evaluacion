## carga de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

rutas = os.listdir(os.path.join("data"))
df = pd.read_csv(os.path.join("data", rutas[0]))
df = df.iloc[:,1:] #eliminamos la marca temporal
df.head()
```

```python
df.info(verbose=True)
```

## acortamos nombres

```python
nombres_acortados = [
    "Edad", "Genero", "Tipo institucion",
    "Estatus Ac", "Estado del estudiante", "Promedio Actual/Egr",
    "¿Test vocacional?", "Utilidad test", "Fuentes de decisión carrera",
    "Tiempo inv de carreras", "Dispositivo de investigacion", "Acceso a internet",
    "Plataformas Inv mas usadas", "Formatos de Orientación", "Interes en la app",
    "Confianza en la app", "Beneficio esperado", "Sistema Ideal p/Elección",
    "Frustración en Elección", "Disposición a Test", "Confianza en la Ayuda",
    "Características Sistema Ideal", "Otra plataforma", "Cual plataforma",
    "Satisfaccion otra plataforma", "Problema tests actuales"
]
pd.DataFrame(df.columns)
```

```python
df_nombres_acortados = df.copy()
df_nombres_acortados.columns = nombres_acortados
df_nombres_acortados.head()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_nombres_acortados, x='Edad', palette='Blues_d')
plt.title('Distribución de Edad de los Encuestados')
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.savefig('edad_distribucion.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
genero_counts = df_nombres_acortados['Genero'].value_counts()
plt.pie(genero_counts, labels=genero_counts.index, autopct='%1.1f%%', 
        startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribución por Género')
plt.savefig('genero_distribucion.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df_nombres_acortados, x='Tipo institucion', palette='Set2')
plt.title('Tipo de Institución de los Encuestados')
plt.xlabel('Tipo de Institución')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.savefig('tipo_institucion.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_nombres_acortados, x='¿Test vocacional?', 
              palette='viridis', order=df_nombres_acortados['¿Test vocacional?'].value_counts().index)
plt.title('Experiencia Previa con Tests Vocacionales')
plt.xlabel('¿Ha usado tests vocacionales?')
plt.ylabel('Frecuencia')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('experiencia_tests.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
utilidad_data = df_nombres_acortados['Utilidad test'].dropna()
sns.histplot(utilidad_data, bins=5, kde=False, color='coral', edgecolor='black')
plt.title('Utilidad Percibida de Tests Vocacionales Previos')
plt.xlabel('Calificación de Utilidad (1-5)')
plt.ylabel('Frecuencia')
plt.xticks([1, 2, 3, 4, 5])
plt.tight_layout()
plt.savefig('utilidad_tests.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_nombres_acortados, y='Fuentes de decisión carrera', 
              order=df_nombres_acortados['Fuentes de decisión carrera'].value_counts().index,
              palette='muted')
plt.title('Fuentes de Información para Decisión de Carrera')
plt.xlabel('Frecuencia')
plt.ylabel('Fuente')
plt.tight_layout()
plt.savefig('fuentes_decision.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_nombres_acortados, x='Dispositivo de investigacion',
              order=df_nombres_acortados['Dispositivo de investigacion'].value_counts().index,
              palette='coolwarm')
plt.title('Dispositivo Más Utilizado para Investigar')
plt.xlabel('Tipo de Dispositivo')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.savefig('dispositivo_investigacion.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_nombres_acortados, x='Acceso a internet',
              order=df_nombres_acortados['Acceso a internet'].value_counts().index,
              palette='Greens_d')
plt.title('Tipo de Acceso a Internet')
plt.xlabel('Acceso a Internet')
plt.ylabel('Frecuencia')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('acceso_internet.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
interes_counts = df_nombres_acortados['Interes en la app'].value_counts()
colors = ['#ff9999', '#ffcc99', '#99ff99']
plt.bar(interes_counts.index, interes_counts.values, color=colors, edgecolor='black')
plt.title('Nivel de Interés en la Aplicación')
plt.xlabel('Respuesta')
plt.ylabel('Frecuencia')
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig('interes_app.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
confianza_data = df_nombres_acortados['Confianza en la app'].dropna()
sns.histplot(confianza_data, bins=5, kde=False, color='steelblue', edgecolor='black')
plt.title('Nivel de Confianza en Sistemas de IA')
plt.xlabel('Nivel de Confianza (1-5)')
plt.ylabel('Frecuencia')
plt.xticks([1, 2, 3, 4, 5])
plt.tight_layout()
plt.savefig('confianza_ia.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_nombres_acortados, x='Edad', y='Confianza en la app', 
               hue='¿Test vocacional?', split=True)
plt.title('Confianza en IA según Edad y Experiencia Previa con Tests')
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=df_nombres_acortados, x='Dispositivo de investigacion', 
              hue='Tiempo inv de carreras', palette='viridis')
plt.title('Dispositivo Usado vs Tiempo Invertido en Investigación')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

matriz_satisfaccion = df_nombres_acortados[['Utilidad test', 
                                             'Satisfaccion otra plataforma', 
                                             'Confianza en la app']].dropna()
sns.heatmap(matriz_satisfaccion.corr(), annot=True, cmap='coolwarm', ax=ax[0])
ax[0].set_title('Correlación: Satisfacción y Confianza')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=30, ha='right')

sns.boxplot(data=df_nombres_acortados, x='Utilidad test', 
            y='Confianza en la app', ax=ax[1])
ax[1].set_title('Utilidad Percibida vs Confianza Futura')
```

```python
plt.figure(figsize=(14, 7))
frustracion_orden = df_nombres_acortados['Frustración en Elección'].value_counts().index
sns.countplot(data=df_nombres_acortados, y='Frustración en Elección', 
              hue='Disposición a Test', order=frustracion_orden, palette='rocket')
plt.title('Tipo de Frustración vs Disposición a Usar Test Digital')
```

```python
sns.countplot(data=df_nombres_acortados, x="Tipo institucion", hue="Interes en la app")
plt.title("Interés en la App según el Tipo de Institución")
plt.show()
```
